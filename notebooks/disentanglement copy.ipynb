## Soft labels
Here we would expect aleatoric uncertainty to increase while epistemic stays similar.

```python
def read_uncertainties(path: Path, percentile: int) -> tuple[np.ndarray, np.ndarray]:
    """Load all AU and EU uncertainties per token for a given dataset."""
    if not path.exists():
        raise ValueError(path)

    df = pd.read_json(path, lines=True)

    all_au_flat = np.concatenate(df["all_au"])
    all_mi_flat = np.concatenate(df["all_mi"])

    total = all_au_flat + all_mi_flat
    threshold = np.percentile(total, percentile)

    mask = total > threshold
    print(f"setting threshold at {threshold} to keep {100 * mask.sum() / len(mask)}% ({mask.sum()} out of {len(mask)})")

    return all_au_flat[mask], all_mi_flat[mask]

def load_results(method: str, dataset: str, percentile: int) -> dict[str, np.ndarray]:
    """Load AU and EU uncertainties for hard and soft labels given a specific method and dataset."""
    results = {}
    for infix in ("hard", "soft"):
        path = Path("../results") / dataset / f"{method}-{infix}-token.jsonl"
        all_au, all_mi = read_uncertainties(path, percentile)
        results[f"{infix}-au"] = all_au
        results[f"{infix}-mi"] = all_mi

    return results

def draw(results: dict) -> None:
    df_au = pd.DataFrame({
        "Value": list(results["soft-au"]) + list(results["hard-au"]),
        "Condition": ["Soft"] * len(results["soft-au"]) + ["Hard"] * len(results["hard-au"])
    })
    df_mi = pd.DataFrame({
        "Value": list(results["soft-mi"]) + list(results["hard-mi"]),
        "Condition": ["Soft"] * len(results["soft-mi"]) + ["Hard"] * len(results["hard-mi"])
    })

    _, axs = plt.subplots(1, 2, figsize=(12, 5), tight_layout=True)

    sns.violinplot(data=df_au, x="Condition", y="Value", hue="Condition", inner="box", palette="Set2", legend=False, ax=axs[0])
    axs[0].set_ylabel("Uncertainty")

    sns.violinplot(data=df_mi, x="Condition", y="Value", hue="Condition", inner="box", palette="Set2", legend=False, ax=axs[1])
    axs[0].set_title("Aleatoric Uncertainty (AU)")
    axs[1].set_title("Epistemic Uncertainty (MI)")
    axs[1].set_ylabel("")

    # plt.suptitle("Uncertainty Disentanglement: Hard vs Soft Labels")
    plt.show()


def perform_test(results: dict) -> None:
    for key, label in [("au", "AU"), ("mi", "MI")]:
        soft = results[f"soft-{key}"]
        hard = results[f"hard-{key}"]

                # we hypothesize that for AU soft should be greater
                #                 and for EU soft should not change
        stat, p = mannwhitneyu(soft, hard, alternative="greater" if key == "au" else "two-sided")
        r = 1 - (2 * stat) / (len(soft) * len(hard))

        soft_median = np.median(soft)
        hard_median = np.median(hard)

        print(f"{label}: median soft={soft_median:.3f}, hard={hard_median:.3f}, p={p:.4e}, effect size r={r:.3f}")


def make_hists() -> None:
    methods = ("ensemble5",)
    datasets = ("webquestions","nq")
    for method in methods:
        for dataset in datasets:
            print("CHECKING DATASET", dataset)
            for percentile in range(0, 55, 5):
                results = load_results(method, dataset, percentile)
                draw(results)
                perform_test(results)
                print()

make_hists()
```

```python
"""Paired per-input (T, omega) readout for the soft-vs-hard-label experiment.

Replaces the pooled token-level comparison:
- aggregates token uncertainties per input (ratio of sums, so dilution cancels in omega)
- pairs inputs across arms and computes (d log T, d logit omega)
- Wilcoxon signed-rank with pre-registered one-sided predictions
- displacement scatter + paired-difference violins
- robustness: composition shift within the high-T half (stratified on the
  baseline arm's T; characterizes noise-robustness, makes no claim about r(A,E))

Caveats to carry into the write-up:
- one training run per arm: p-values describe this pair of runs, not the
  intervention over training stochasticity (seeds would be needed for that)
- unweighted first pass: precision weights 1/s2_i need per-mask logs
- provisional until the per-arm recalibration check
"""

from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import wilcoxon

EPS = 1e-9  # floor for logs; clipped inputs are dropped and counted
ALIGNMENT_KEYS = ("question", "id", "qid")  # first one present is used to verify pairing


def read_per_input(path: Path) -> pd.DataFrame:
    """Per-input aggregates from token-level logs. No pooling across inputs."""
    if not path.exists():
        raise FileNotFoundError(path)
    df = pd.read_json(path, lines=True)
    au_sum = df["all_au"].apply(np.sum).to_numpy(dtype=float)
    mi_sum = df["all_mi"].apply(np.sum).to_numpy(dtype=float)
    out = pd.DataFrame(
        {
            "au_sum": au_sum,
            "mi_sum": mi_sum,
            "t_sum": au_sum + mi_sum,
            "n_tokens": df["all_au"].apply(len).to_numpy(),
        }
    )
    for key in ALIGNMENT_KEYS:
        if key in df.columns:
            out[key] = df[key].values
            break
    out["log_t"] = np.log(np.maximum(out["t_sum"], EPS))
    # omega from the ratio of sums; logit omega = log(A/E)
    out["logit_omega"] = np.log(np.maximum(out["au_sum"], EPS)) - np.log(
        np.maximum(out["mi_sum"], EPS)
    )
    out["clipped"] = (out["au_sum"] < EPS) | (out["mi_sum"] < EPS)
    return out


def load_paired(method: str, dataset: str, results_dir: Path = Path("../results")) -> pd.DataFrame:
    arms = {}
    for infix in ("hard", "soft"):
        path = results_dir / dataset / f"{method}-{infix}-token.jsonl"
        arms[infix] = read_per_input(path)
    hard, soft = arms["hard"], arms["soft"]

    if len(hard) != len(soft):
        raise ValueError(f"arms differ in length ({len(hard)} vs {len(soft)}); check eval sets")
    key = next((k for k in ALIGNMENT_KEYS if k in hard.columns and k in soft.columns), None)
    if key is not None:
        if not (hard[key].values == soft[key].values).all():
            raise ValueError(f"row order differs between arms on '{key}'; merge on the key instead")
    else:
        print("WARNING: no alignment key found; pairing assumes identical row order in both files")

    paired = pd.DataFrame(
        {
            "d_log_t": soft["log_t"].values - hard["log_t"].values,
            "d_logit_omega": soft["logit_omega"].values - hard["logit_omega"].values,
            "log_t_hard": hard["log_t"].values,  # baseline-arm scale, for stratification
            "clipped": soft["clipped"].values | hard["clipped"].values,
        }
    )
    n_clip = int(paired["clipped"].sum())
    if n_clip:
        print(f"dropping {n_clip} inputs with floored AU or MI sums ({100 * n_clip / len(paired):.1f}%)")
    return paired.loc[~paired["clipped"]].reset_index(drop=True)


def rank_biserial(d: np.ndarray) -> float:
    """Matched-pairs rank-biserial correlation for the signed-rank test."""
    d = d[d != 0]
    ranks = pd.Series(np.abs(d)).rank().to_numpy()
    r_plus = ranks[d > 0].sum()
    r_minus = ranks[d < 0].sum()
    return (r_plus - r_minus) / (r_plus + r_minus)


def test_paired(paired: pd.DataFrame) -> None:
    predictions = [
        ("d_logit_omega", "greater", "Dlogit omega (pred: soft > hard)"),
        ("d_log_t", "greater", "Dlog T     (pred: soft > hard)"),
    ]
    for col, alt, label in predictions:
        d = paired[col].to_numpy()
        stat, p = wilcoxon(d, alternative=alt)
        print(
            f"{label}: median D={np.median(d):+.4f}, frac>0={(d > 0).mean():.3f}, "
            f"rank-biserial r={rank_biserial(d):+.3f}, n={len(d)}, p={p:.3e}"
        )


def high_t_robustness(paired: pd.DataFrame) -> None:
    """Composition shift within the high-T half, stratified on the BASELINE arm.

    This characterizes noise-robustness of the scale-free coordinate; it is not
    the prohibited conditioning (no claim about A-E dependence is read off).
    """
    hi = paired.loc[paired["log_t_hard"] >= paired["log_t_hard"].median()]
    stat, p = wilcoxon(hi["d_logit_omega"], alternative="greater")
    print(
        f"high-T half: median Dlogit omega={hi['d_logit_omega'].median():+.4f}, "
        f"n={len(hi)}, p={p:.3e}"
    )


def draw_displacement(paired: pd.DataFrame, title: str) -> None:
    fig, axs = plt.subplots(1, 2, figsize=(12, 5), tight_layout=True)

    axs[0].scatter(paired["d_log_t"], paired["d_logit_omega"], s=8, alpha=0.35)
    axs[0].axhline(0, lw=0.8, color="gray")
    axs[0].axvline(0, lw=0.8, color="gray")
    axs[0].set_xlabel(r"$\Delta \log T$ (soft $-$ hard)")
    axs[0].set_ylabel(r"$\Delta \mathrm{logit}\,\omega$ (soft $-$ hard)")
    axs[0].set_title("Per-input displacement")

    melted = paired.melt(value_vars=["d_log_t", "d_logit_omega"], var_name="coord", value_name="delta")
    sns.violinplot(data=melted, x="coord", y="delta", hue="coord", inner="box",
                   palette="Set2", legend=False, ax=axs[1])
    axs[1].axhline(0, lw=0.8, color="gray")
    axs[1].set_xticklabels([r"$\Delta \log T$", r"$\Delta \mathrm{logit}\,\omega$"])
    axs[1].set_ylabel("paired difference")
    axs[1].set_title("Paired differences")

    fig.suptitle(title)
    plt.show()


def run(method: str = "ensemble5", datasets: tuple[str, ...] = ("webquestions", "nq")) -> None:
    for dataset in datasets:
        print(f"=== {method} / {dataset} ===")
        paired = load_paired(method, dataset)
        test_paired(paired)
        high_t_robustness(paired)
        draw_displacement(paired, f"{method} / {dataset}")
        print()


if __name__ == "__main__":
    run()
```

# Training on subsets

```python
def load_results_subset(method: str, dataset: str, percentile: int) -> dict[str, np.ndarray]:
    results = {}
    # full model
    path_full = Path("../results") / dataset / f"{method}-hard-token.jsonl"
    results["hard-au"], results["hard-mi"] = read_uncertainties(path_full, percentile)
    # subset model (0.5)
    path_subset = Path("../results") / dataset / f"{method}-hard-0.1-token.jsonl"  # adjust filename
    results["subset-au"], results["subset-mi"] = read_uncertainties(path_subset, percentile)
    return results


def perform_subset_test(results: dict) -> None:
    for key, label in [("au", "AU"), ("mi", "MI")]:
        subset = results[f"subset-{key}"]
        full = results[f"hard-{key}"]
        # hypothesis: less data -> higher EU (one-sided), AU should not change (two-sided)
        alternative = "greater" if key == "mi" else "two-sided"
        stat, p = mannwhitneyu(subset, full, alternative=alternative)
        r = 1 - (2 * stat) / (len(subset) * len(full))
        print(f"{label}: median subset={np.median(subset):.3f}, full={np.median(full):.3f}, p={p:.4e}, r={r:.3f}")


def draw_subset(results: dict, dataset: str) -> None:
    df_au = pd.DataFrame({
        "Value":     list(results["subset-au"]) + list(results["hard-au"]),
        "Condition": ["50% data"] * len(results["subset-au"]) + ["100% data"] * len(results["hard-au"])
    })
    df_mi = pd.DataFrame({
        "Value":     list(results["subset-mi"]) + list(results["hard-mi"]),
        "Condition": ["50% data"] * len(results["subset-mi"]) + ["100% data"] * len(results["hard-mi"])
    })
    _, axs = plt.subplots(1, 2, figsize=(12, 5), tight_layout=True)
    sns.violinplot(data=df_au, x="Condition", y="Value", hue="Condition",
                   inner="box", palette="Set2", legend=False, ax=axs[0])
    sns.violinplot(data=df_mi, x="Condition", y="Value", hue="Condition",
                   inner="box", palette="Set2", legend=False, ax=axs[1])
    axs[0].set_title(f"AU — {dataset}")
    axs[1].set_title(f"EU (MI) — {dataset}")
    axs[0].set_ylabel("Uncertainty")
    axs[1].set_ylabel("")
    plt.show()


for dataset in ("nq",):
    for percentile in range(0, 5, 5):
        results = load_results_subset("ensemble5", dataset, percentile)
        draw_subset(results, dataset)
        perform_subset_test(results)
        print()
# import sys
# sys.path.insert(0, "..")

# from src.eval.utils import exact_match


# path = Path("../results") / "nq" / "ensemble5-hard-0.1-token.jsonl"
# df = pd.read_json(path, lines=True)
# df["em"] = df.apply(lambda r: exact_match(r["prediction"], r["answers"]), axis=1)
# print(df["em"].mean())
```

```python
from scipy.stats import pearsonr

import matplotlib.pyplot as plt
import numpy as np
from statsmodels.nonparametric.smoothers_lowess import lowess

def plot_t_ratio(T, ratio, color_by=None, save_path=None, log_x=False):
    """
    Scatter of T vs A/T with marginal histograms and a LOESS trend.
    color_by: optional array (e.g. binary correct/incorrect) for point colour.
    """
    fig = plt.figure(figsize=(6.5, 5.5))
    gs = fig.add_gridspec(
        2, 2,
        width_ratios=(5, 1), height_ratios=(1, 5),
        wspace=0.04, hspace=0.04,
    )
    ax = fig.add_subplot(gs[1, 0])
    ax_top = fig.add_subplot(gs[0, 0], sharex=ax)
    ax_right = fig.add_subplot(gs[1, 1], sharey=ax)

    # main scatter
    if color_by is not None:
        sc = ax.scatter(T, ratio, c=color_by, alpha=0.4, s=8,
                        cmap='coolwarm', rasterized=True)
    else:
        ax.scatter(T, ratio, alpha=0.25, s=6,
                   color='#3a6fa3', rasterized=True)

    # LOESS trend, robust to outliers
    sm = lowess(ratio, T, frac=0.25, it=2, return_sorted=True)
    ax.plot(sm[:, 0], sm[:, 1], color='crimson', lw=2, label='LOESS')

    # marginals
    ax_top.hist(T, bins=60, color='#3a6fa3', alpha=0.8)
    ax_right.hist(ratio, bins=60, orientation='horizontal',
                  color='#3a6fa3', alpha=0.8)

    # cosmetics
    if log_x:
        ax.set_xscale('log')
    ax.set_ylim(0, 1)
    ax.set_xlabel(r'Total uncertainty $T$')
    ax.set_ylabel(r'Aleatoric fraction $A/T$')
    ax.legend(loc='best', frameon=False, fontsize=9)

    for a in (ax_top, ax_right):
        a.spines[['top', 'right']].set_visible(False)
    ax_top.spines['left'].set_visible(False)
    ax_right.spines['bottom'].set_visible(False)
    ax_top.tick_params(left=False, labelleft=False, labelbottom=False)
    ax_right.tick_params(bottom=False, labelbottom=False, labelleft=False)
    ax.spines[['top', 'right']].set_visible(False)

    # plt.tight_layout()
    # if save_path:
        # plt.savefig(save_path, dpi=300, bbox_inches='tight')
    return fig, ax


def read_uncertainties2(path: Path) -> tuple[np.ndarray, np.ndarray]:
    """Load all AU and EU uncertainties per token for a given dataset."""
    if not path.exists():
        raise ValueError(path)

    df = pd.read_json(path, lines=True)

    all_au_flat = np.concatenate(df["all_au"])
    all_mi_flat = np.concatenate(df["all_mi"])
    all_tu_flat = all_au_flat + all_mi_flat
    all_ratio_flat = all_au_flat / (all_tu_flat + 1e-8)
    epsilon = 1.0
    mask = all_tu_flat > epsilon

    print("r:", pearsonr(all_au_flat[mask], all_mi_flat[mask]))

    # print("r:", pearsonr(all_tu_flat[mask], all_ratio_flat[mask]))

    # var_a = np.var(all_au_flat)
    # var_e = np.var(all_mi_flat)
    # var_t = np.var(all_tu_flat)

    # print("Var A:", var_a)
    # print("Var E:", var_e)
    # print("Var T:", var_t)
    # print("r:", pearsonr(all_au_flat, all_mi_flat).statistic)
    # up = var_t - var_e - var_a
    # down = 2 * np.sqrt(var_a * var_e)
    # print(up / down)

    plt.figure(figsize=(6, 5))
    plt.scatter(all_au_flat[mask], all_mi_flat[mask], alpha=0.25, s=6, color='#3a6fa3', rasterized=True)
    title = " ".join(str(path).split("/")[2:]).split("-")[0]
    K = 5

    plt.axhline(
        y=np.log(K),
        color='black', linestyle='--', linewidth=1.2,
        label=f'$E \\leq \\log K = {np.log(K):.2f}$'
    )

    plt.title(title)
    plt.xlabel(r"Aleatoric uncertainty $A$")
    plt.ylabel(r"Epistemic uncertainty $E$")
    plt.ylim(0, max(all_mi_flat[mask].max(), np.log(K)) * 1.05)
    plt.legend(frameon=False, loc='best')
    plt.gca().spines[['top', 'right']].set_visible(False)
    plt.tight_layout()
    plt.show()

    plot_t_ratio(all_tu_flat[mask], all_ratio_flat[mask])
    plt.suptitle(title)

    plt.show()

datasets = ("webquestions", "nq")
df_paths = ("mcdropout-hard-token.jsonl",)
for dataset in datasets:
    for df_path in df_paths:
        print(dataset, df_path)
        path = Path("../results") / dataset / df_path

        read_uncertainties2(path)
        print()

```
